```python
import folium

# Define the center of the map and zoom level
center = [49.6116, 6.1319]
zoom = 10

# Create a map object with the specified center and zoom level
m = folium.Map(location=center, zoom_start=zoom)

# Add markers for the main cities in Luxembourg
marker1 = folium.Marker([49.6116, 6.1319], popup="Luxembourg City")
marker2 = folium.Marker([49.6278, 6.1597], popup="Esch-sur-Alzette")
marker3 = folium.Marker([49.5049, 5.9486], popup="Dudelange")
marker4 = folium.Marker([49.6100, 6.1379], popup="Kirchberg")
marker5 = folium.Marker([49.6933, 6.4383], popup="Ettelbruck")

# Add the markers to the map
marker1.add_to(m)
marker2.add_to(m)
marker3.add_to(m)
marker4.add_to(m)
marker5.add_to(m)

# Display the map
m
```
